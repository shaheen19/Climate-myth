```python
#import dependencies
import pandas as pd
import requests as req
#create separate file with individual log in credentials to be imported

from config import (user_id_aqs, password_aqs, Plotkey, Plotuser)
import csv
import matplotlib.pyplot as plt
```

```python
codes = []
countycodes = ["{0:03}".format(i) for i in range(117) if i%2 > 0]

for countycode in countycodes:
    codes.append(countycode)
#codes
```

```python
#API REQUEST
URL_new = f"https://aqs.epa.gov/api/rawData?user={user_id_aqs}&pw={password_aqs}&format=DMCSV&param=88101&bdate=20140101&edate=20141231&state=06&county="
list_2014 = []

for x in codes:
    url_loop = URL_new + x
    print(f"fetching county code {x}")
    response1 = req.get(url_loop)
    print(f"Got API Response for {x}")
    wrapper = csv.reader(response1.text.strip().split('\n'))
    header = next(wrapper)
    for record in wrapper:
        list_2014.append(record)
    print(f"Appended List with {x} API Response")   
        
```

```python
#Create DF and append FIPS column with county code and 06 (state code) to create FIPS code
data_2014 = pd.DataFrame(list_2014, columns = header)
data_2014["FIPS"] = "06" + data_2014["County Code"]
data_2014
```

```python
countyname = { "001": "Alameda", "003": "Apline", "005": "Amador", "007":"Butte", "009":"Calaveras", "011": "Colusa", "013":"Contra Costa", "015":"Del Norte", "017":"El Dorado", "019": "Fresno", "021": "Glenn", "023": "Humboldt", "025": "Imperial", "027": "Inyo", "029": "Kern", "031": "Kings", "033":"Lake", "035": "Lassen", "037": "Los Angeles", "039": "Madera", "041": "Marin", "043": "Mariposa", "045": "Mendicino", "047": "Merced", "049": "Modoc", "051": "Mono", "053": "Monterey", "055": "Napa", "057": "Nevada", "059": "Orange", "061": "Placer", "063": "Plumas", "065": "Riverside", "067": "Sacramento", "069": "San Benito", "071": "San Bernardino", "073": "San Diego", "075": "San Francisco", "077": "San Joaquin", "079": "San Luis Obispo", "081": "San Mateo", "083": "Santa Barbara", "085": "Santa Clara", "087": "Santa Cruz", "089": "Shasta", "091": "Sierra", "093": "Siskiyou", "095": "Solano", "097": "Sonoma", "099": "Stanislaus", "101": "Sutter", "103": "Tehama", "105": "Trinity", "107": "Tulare", "109": "Tuolumne", "111": "Ventura", "113": "Yolo", "115": "Yuba"}

df_trim = data_2014[['County Code','Date Local','24 Hour Local','Date GMT','Sample Measurement', "Sample Frequency",'Units of Measure','Sample Duration', "Latitude", "Longitude", "FIPS"]]
df_trim["County Names"] = df_trim["County Code"].map(countyname)
df_trim
```

```python
#Create 2 dataframes to hold records for sampling done by 24 hour duration and 1 hour duration
hour1_df = df_trim.loc[(df_trim["Sample Duration"]== "1 HOUR")]
hour24_df = df_trim.loc[(df_trim["Sample Duration"]== "24 HOUR")]

#See which sampling duration has more unique counties
hour1_df["County Names"].nunique()
hour24_df["County Names"].nunique()
```

```python
#Convert 'Date Local' column to datetime to extract month #
hour1_df['Date Local'] = pd.to_datetime(hour1_df['Date Local'])

#Creating column for Month # for 1hour sample duration df
hour1_df['Month'] = pd.DatetimeIndex(hour1_df['Date Local']).month
hour1_df["Month"].fillna(0)
hour1_df["Month"].astype(float)
hour1_df.fillna(0)
hour1_df["County Names"].dtype

#Convert 'Date Local' column to datetime to extract month #
hour24_df['Date Local'] = pd.to_datetime(hour24_df['Date Local'])

#Creating column for Month # for 24 hour sample duration df
hour24_df['Month'] = pd.DatetimeIndex(hour24_df['Date Local']).month
hour24_df["Month"].fillna(0)
hour24_df["Month"].astype(float)
hour24_df.fillna(0)
hour24_df.head()
```

```python
#extract only relevant columns and create a groupby df (for 24 hour sample duration)
mday_df = hour24_df.loc[:, ['County Names', 'Month', "Sample Measurement", "Latitude", "Longitude", "FIPS"]]
mday_df["Sample Measurement"].astype(float)

#Convert column dtype: "O" to numeric for groupby functionality 
tuff24 = pd.to_numeric(mday_df["Sample Measurement"], errors='coerce')
tuff24
mday_df["PM2.5 Values 2014"] = tuff24
finalday = mday_df
finalday.drop(columns="Sample Measurement")
#finalday
finalday.to_csv("2014CA24hrPM2.5.csv")
finalday.head()
```

```python
#groupby to graph results
county_date24hr = pd.DataFrame(finalday.groupby(['County Names', 'Month']).mean())
county_date24hr.head()
```

```python
county_date24hr.unstack(0).plot(figsize = (10,10))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title("Graph of PM2.5 Monthly Averages in CA Counties 2014 (24hr Sampling Duration)")
plt.ylabel("PM2.5 Mean Per Month")
plt.xlabel("Month")
plt.savefig("2014 PM2.5 AVG (24hr Sampling).png", bbox_inches='tight')
plt.figure()
```

```python
# extract only relevant columns and create a groupby df
mhour_df = hour1_df.loc[:, ['County Names', 'Month', "Sample Measurement", "Latitude", "Longitude", "FIPS"]]
mhour_df["Sample Measurement"].astype(float)

#Convert column dtype: "O" to numeric for groupby functionality 
tuffhour_df = pd.to_numeric(mhour_df["Sample Measurement"], errors='coerce')
tuffhour_df
mhour_df["PM2.5 Values 2014"] = tuffhour_df
finalhour = mhour_df
finalhour.drop(columns="Sample Measurement")

#Convert column dtype: "O" to numeric for groupby functionality 
tuffhour_df = pd.to_numeric(mday_df["Sample Measurement"], errors='coerce')
tuffhour_df
mday_df["PM2.5 Values 2014"] = tuffhour_df
finalhour = mday_df
finalhour.drop(columns="Sample Measurement")

finalhour.to_csv("2014CA1hrPM2.5.csv")
```

```python
#groupby to graph results
county_1hr = pd.DataFrame(finalhour.groupby(['County Names', 'Month']).mean())
county_1hr.head()
```

```python
county_1hr.unstack(0).plot(figsize = (10,10))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title("Graph of PM2.5 Monthly Averages in CA Counties 2014 (1Hr Sampling duration)")
plt.ylabel("PM2.5 Mean Per Month")
plt.xlabel("Month")
plt.savefig("2014 PM2.5 AVG (1hrSampling).png", bbox_inches='tight')
plt.figure()        
```

```python
#Find Min
min_pm25 = county_date24hr['PM2.5 Values 2014'].min()
min_county24hr = county_date24hr.loc[county_date24hr['PM2.5 Values 2014'] == min_pm25,:]
min_county24hr

min_pm25 = county_1hr['PM2.5 Values 2014'].min()
min_county1hr = county_1hr.loc[county_1hr['PM2.5 Values 2014'] == min_pm25,:]
min_county1hr
```

```python
# create a df with only the Max & Min, and SD/LA for ref counties
minmax_df = finalday.loc[(finalday['County Names']=='Tulare') | (finalday['County Names']=='Lake') | (finalday['County Names']=='San Diego') | (finalday['County Names']=='Los Angeles'),:]
minmax_df.head()
```

```python
mm_mean_df = minmax_df.loc[:, ['County Names', 'Month', 'PM2.5 Values 2014']]
mm_df = pd.DataFrame(mm_mean_df.groupby(['County Names', 'Month']).mean())
mm_df.head()
```

```python
mm_df.unstack(0).plot(figsize = (10,10))
plt.title("Graph of PM2.5 Max & Min in CA + San Diego & Los Angeles 2014")
plt.xlabel('Month')
plt.ylabel('PM2.5 Mean/Month')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 0.5))
plt.savefig("2014_CA_MaxMin_SDLA.png", bbox_inches='tight') 
plt.figure()
```

```python
import plotly
import plotly.plotly as py
import plotly.figure_factory as ff
plotly.tools.set_credentials_file(username=Plotuser, api_key=Plotkey)
```

```python
#create lists of values to graph in plotly
values = county_sum["PM2.5 Values 2014"].tolist()
fips = county_sum.index.values
print(values)
```

```python
#send FIPS codes and corresponding PM2.5 values to plotly to graph
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['CA'],
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='PM2.5 by County', title='California PM2.5 2014 By County'
)
py.iplot(fig, filename='CA_PM2.5 2014')
```
