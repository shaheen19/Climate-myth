```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
asthma_df = pd.read_csv('Resources/asthma-ed-visit-rates-lghc-indicator-07.csv')
asthma_df.head(10)
```

```python
filter_out_ca = asthma_df.loc[asthma_df['Geography'] != 'California', :]
filter_out_ca
```

```python
group = filter_out_ca.groupby(['Geography', 'Year'])['Rate'].mean()
```

```python
rates_df = pd.DataFrame(group).unstack(0)
```

```python
rates_df.head()
```

```python
rates_df.columns
```

```python
plt.plot(rates_df)
plt.xlabel('Year')
plt.ylabel('ED Visit Rates')
#plt.legend(rates_df[''])
```
