```python
import pandas as pd
```

```python
# The path to our CSV file
file = "Resources/asthma-visits.csv"

# Read our Kickstarter data into pandas
df = pd.read_csv(file)
df.head()
```

```python
year = df["Year"]
```

```python
#rate = rate per 10,000 residents
gs = df.groupby(['Geography', year])['Rate'].mean().unstack()#.fillna(0)
gs
```

```python
gsc = gs.fillna(0)
gsc.count()
```

```python
gsg = df.groupby(['Strata Name', year])['Rate'].mean().unstack()#.fillna(0)
gsg
```

```python
df['Numerator'] = pd.to_numeric(df['Numerator'], errors='coerce')

gs2 = df.groupby(['Geography', year])['Numerator'].mean().unstack()#.fillna(0)
gs2
```
